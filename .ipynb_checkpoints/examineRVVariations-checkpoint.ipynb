```python
TCE_ID = []
for i in range(len(cfpTable)):
    TCE_ID.append("%09d"%cfpTable.kepid.loc[i] + "-%02d"%cfpTable.koi_tce_plnt_num.loc[i])

cfpTable["TCE_ID"] = TCE_ID
```

```python
dataLoc = "../../../../occurenceRates/DR25OccurrencePynb/data/"

hiCompletenessTces = ascii.read(dataLoc + "kplr_dr25_obs_highcompleteness_tces.txt");
hiCompletenessTces = hiCompletenessTces.to_pandas();

fpwgpcTces = ascii.read(dataLoc + "kplr_dr25_obs_fpwgpc_tces.txt");
fpwgpcTces = fpwgpcTces.to_pandas();

dr25Tces = ascii.read(dataLoc + "kplr_dr25_obs_tces.txt");
dr25Tces = dr25Tces.to_pandas();
```

```python
hiCompletenessTcesIsPC = hiCompletenessTces[hiCompletenessTces["Disp"] == "PC"]
print(str(len(hiCompletenessTcesIsPC)) + " high completeness TCEs are PCs")
hiCompletenessTcesIsPCNotKoi = hiCompletenessTcesIsPC[ ~hiCompletenessTcesIsPC['TCE_ID'].isin(cfpTable['TCE_ID'])]
print(str(len(hiCompletenessTcesIsPCNotKoi)) + " high completeness TCE PCs are not KOIs")

fpwgpcTcesIsPC = fpwgpcTces[fpwgpcTces["Disp"] == "PC"]
print(str(len(fpwgpcTcesIsPC)) + " FPWG TCEs are PCs")
fpwgpcTcesIsPCNotKoi = fpwgpcTcesIsPC[ ~fpwgpcTcesIsPC['TCE_ID'].isin(cfpTable['TCE_ID'])]
print(str(len(fpwgpcTcesIsPCNotKoi)) + " FPWG TCE PCs are not KOIs")

```

```python
import matplotlib.patches as patches
cbc = ['#377eb8', '#ff7f00', '#4daf4a',
                  '#f781bf', '#a65628', '#984ea3',
                  '#999999', '#e41a1c', '#dede00']
fig, ax = plt.subplots(figsize=(10,6));
scf = plt.scatter(fpwgpcTcesIsPCNotKoi.period, fpwgpcTcesIsPCNotKoi.Rp, color=cbc[0], label="Non-KOI FPWG PCs");
scf = plt.scatter(hiCompletenessTcesIsPCNotKoi.period, hiCompletenessTcesIsPCNotKoi.Rp, 
                 s = 100, marker='s', facecolors="none", edgecolors=cbc[1], label="Non-KOI High Completeness PCs");
plt.legend()
plt.xlim(0, 740)
xl = plt.xlim()
yl = plt.ylim()
# plt.plot([300, 300],[0, 2],'k',lw=0.6,alpha=0.5)
# plt.plot([300, xl[1]],[2, 2],'k',lw=0.6,alpha=0.5)
rect = patches.Rectangle((80,0), xl[1]-80, yl[1], facecolor = 'k', alpha = 0.08)
plt.gca().add_patch(rect)

plt.xlabel("period", fontsize = 24);
plt.ylabel("planet radius", fontsize = 24);
plt.ylim([0, 20])
plt.tick_params(labelsize = 16)
plt.savefig("newNonKoiPCs.pdf",bbox_inches='tight')
# plt.xlim([50, 400])

# cbh = plt.colorbar(scf);
# cbh.ax.set_ylabel("Reliability");
plt.title("KOI Reliability, size = total reliability");
```

```python
durOut = 2.904/24
durOutSigma = 0.5607/24
perOut = 5.533317
durIn = 3.638/24
durInSigma = 0.5113/24
perIn = 3.422508

xiUp = durIn / (perIn**(1/3))
xiUpSigma = durInSigma / (perIn**(1/3))

xiDown = durOut / (perOut**(1/3))
xiDownSigma  = durOutSigma / (perOut**(1/3))

xi = xiUp/xiDown
xiSigma = np.sqrt((xiUpSigma**2/xiDown**2) + xiUp**2 * xiDownSigma**2 / xiDown**4)

print([xi, xiSigma])
```

```python
print([np.log10(xi-xiSigma), np.log10(xi), np.log10(xi+xiSigma)])
```
